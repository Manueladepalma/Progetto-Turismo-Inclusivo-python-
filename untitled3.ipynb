```python
# Progetto Turismo Inclusivo in Italia
# Librerie necessarie
import pandas as pd
import matplotlib.pyplot as plt

# Creazione del dataset simulato
data = {
    "Nome": ["Hotel Inclusivo Roma", "B&B Accessibile Firenze", "Agriturismo Senza Barriere Napoli",
             "Hotel Accessibile Milano", "B&B Accogliente Torino"],
    "Posizione": ["Lazio, Roma", "Toscana, Firenze", "Campania, Napoli", "Lombardia, Milano", "Piemonte, Torino"],
    "Servizi": ["Rampe, Ascensore, Bagni accessibili", "Ascensore, Guide tattili",
                "Rampe, Servizi per ipovedenti", "Ascensore, Bagni accessibili", "Rampe, Guide tattili"],
    "Tipo": ["Hotel", "B&B", "Agriturismo", "Hotel", "B&B"],
    "Valutazione": [4.5, 4.7, 4.3, 4.6, 4.4]
}

# Creazione del DataFrame
df = pd.DataFrame(data)

# Pulizia dei dati: eliminazione duplicati e gestione valori mancanti
df_cleaned = df.dropna().drop_duplicates()

# Statistiche descrittive
print("Statistiche Descrittive del Dataset:")
print(df_cleaned.describe())

# Visualizzazione dei dati
# 1. Grafico delle valutazioni medie per tipo di struttura
plt.figure(figsize=(8, 5))
df_cleaned.groupby("Tipo")["Valutazione"].mean().plot(kind="bar", color="skyblue", title="Valutazione Media per Tipo di Struttura")
plt.ylabel("Valutazione Media")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Distribuzione delle strutture per Regione
df_cleaned['Regione'] = df_cleaned['Posizione'].apply(lambda x: x.split(',')[0])
plt.figure(figsize=(8, 5))
df_cleaned['Regione'].value_counts().plot(kind="bar", color="lightgreen", title="Distribuzione delle Strutture per Regione")
plt.ylabel("Numero di Strutture")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Salvataggio del dataset pulito
df_cleaned.to_csv("progetto_turismo_inclusivo.csv", index=False)
print("\nIl file 'progetto_turismo_inclusivo.csv'  stato salvato con successo!")
```
